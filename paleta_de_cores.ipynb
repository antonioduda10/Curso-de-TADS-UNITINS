```python
from fpdf import FPDF

class PDFWithColors(FPDF):
    def color_box(self, r, g, b, label):
        self.set_fill_color(r, g, b)
        self.cell(40, 10, '', 0, 0, 'C', True)  # Caixa com a cor
        self.cell(100, 10, label, 0, 1, 'L')  # Texto ao lado

# Criação do arquivo PDF
pdf = PDFWithColors()

# Adicionando uma página
pdf.add_page()

# Adicionando a imagem à esquerda
pdf.set_font("Arial", 'B', 16)

# Inserir a imagem (ajuste o caminho da imagem)
pdf.image('C:\\Users\\anton\\OneDrive\\Área de Trabalho\\Cursos Duda\\TO GRADUALDO - UNITINS\\2º semestre 02-2024\\5. INTERFACE HUMANO COMPUTADOR - 2 Período Turma 2024-1 TADS\\Etapa 2\\11. Aula 11\\paleta\\download.png', 10, 8, 33)  # (caminho, posição x, posição y, largura)

# Mover o cursor para a direita da imagem e centralizar o título
pdf.set_xy(50, 15)
pdf.cell(120, 10, txt="INTERFACE HUMANO COMPUTADOR - TADS III", ln=True, align='C')

# Voltar a posição normal para adicionar os nomes abaixo da imagem
pdf.ln(25)  # Adiciona espaçamento para não sobrepor a imagem

# Adicionando os nomes do Professor e Aluno alinhados à esquerda
pdf.set_font("Arial", size=12)
pdf.cell(100, 10, txt="Professor: Alexandre Rossini", ln=True, align='L')
pdf.cell(100, 10, txt="Aluno: Antônio Duda Oliveira da Silva", ln=True, align='L')

# Espaçamento extra abaixo do nome do aluno
pdf.ln(10)

# Subtítulo (centralizado)
pdf.set_font("Arial", 'B', 15)
pdf.cell(0, 10, txt="PALETA DE CORES", ln=True, align='C')

# Espaçamento extra abaixo do título
pdf.ln(10)

# Light Mode
pdf.set_font("Arial", 'B', 14)
pdf.cell(0, 10, txt="Light Mode", ln=True, align='C')

# Adicionando cores com caixas coloridas
pdf.set_font("Arial", size=12)
pdf.color_box(255, 255, 255, "Cor de fundo: #FFFFFF (Branco)")
pdf.color_box(0, 51, 102, "Cor de texto: #003366 (Azul Escuro)")
pdf.color_box(255, 204, 0, "Cor de ação/destaque: #FFCC00 (Amarelo Vibrante)")
pdf.color_box(0, 51, 102, "Cor dos botões: #003366 (Azul Institucional)")
pdf.color_box(255, 0, 0, "Cor de erro: #FF0000 (Vermelho)")
pdf.color_box(34, 139, 34, "Cor de sucesso: #228B22 (Verde Escuro)")
pdf.color_box(255, 140, 0, "Cor de aviso: #FF8C00 (Laranja Escuro)")

# Dark Mode
pdf.ln(10)
pdf.set_font("Arial", 'B', 14)
pdf.cell(0, 10, txt="Dark Mode", ln=True, align='C')

# Adicionando cores com caixas coloridas
pdf.set_font("Arial", size=12)
pdf.color_box(0, 34, 68, "Cor de fundo: #002244 (Azul Muito Escuro)")
pdf.color_box(255, 255, 255, "Cor de texto: #FFFFFF (Branco)")
pdf.color_box(255, 204, 0, "Cor de ação/destaque: #FFCC00 (Amarelo Vibrante)")
pdf.color_box(255, 204, 0, "Cor dos botões: #FFCC00 (Amarelo Vibrante)")
pdf.color_box(255, 99, 71, "Cor de erro: #FF6347 (Vermelho Claro)")
pdf.color_box(50, 205, 50, "Cor de sucesso: #32CD32 (Verde Claro)")
pdf.color_box(255, 165, 0, "Cor de aviso: #FFA500 (Laranja Claro)")

# Salvando o arquivo
pdf_output_path_with_colors = "C:\\Users\\anton\\OneDrive\\Área de Trabalho\\Cursos Duda\\TO GRADUALDO - UNITINS\\2º semestre 02-2024\\5. INTERFACE HUMANO COMPUTADOR - 2 Período Turma 2024-1 TADS\\Etapa 2\\11. Aula 11\\paleta\\paleta.pdf"
pdf.output(pdf_output_path_with_colors)

pdf_output_path_with_colors
```

```python
import matplotlib.pyplot as plt

# Definir as cores e suas descrições
cores_light_mode = [
    ("#FFFFFF", "Cor de fundo: Branco"),
    ("#003366", "Cor de texto: Azul Escuro"),
    ("#FFCC00", "Cor de ação/destaque: Amarelo Vibrante"),
    ("#003366", "Cor dos botões: Azul Institucional"),
    ("#FF0000", "Cor de erro: Vermelho"),
    ("#228B22", "Cor de sucesso: Verde Escuro"),
    ("#FF8C00", "Cor de aviso: Laranja Escuro")
]

cores_dark_mode = [
    ("#002244", "Cor de fundo: Azul Muito Escuro"),
    ("#FFFFFF", "Cor de texto: Branco"),
    ("#FFCC00", "Cor de ação/destaque: Amarelo Vibrante"),
    ("#FFCC00", "Cor dos botões: Amarelo Vibrante"),
    ("#FF6347", "Cor de erro: Vermelho Claro"),
    ("#32CD32", "Cor de sucesso: Verde Claro"),
    ("#FFA500", "Cor de aviso: Laranja Claro")
]

def plot_paleta(cores, title):
    fig, ax = plt.subplots(figsize=(6, len(cores) * 0.5))
    ax.set_title(title, fontsize=16, weight='bold')
    
    # Loop sobre as cores para criar retângulos coloridos
    for i, (cor_hex, label) in enumerate(cores):
        ax.add_patch(plt.Rectangle((0, i), 1, 1, color=cor_hex))
        ax.text(1.1, i + 0.5, label, va='center', ha='left', fontsize=12)
    
    # Ajustar os eixos
    ax.set_xlim(0, 3)
    ax.set_ylim(0, len(cores))
    ax.axis('off')
    plt.show()

# Exibir paletas
plot_paleta(cores_light_mode, "Paleta de Cores - Light Mode")
plot_paleta(cores_dark_mode, "Paleta de Cores - Dark Mode")
```
